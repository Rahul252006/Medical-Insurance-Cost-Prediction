```python
#one hot encoding 
x = pd.get_dummies(x,drop_first=True)
x
# as y is already a numerical value so there is no need of encoding
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
print(model.coef_)
```

```python
#prediction
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import math
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
mse = mean_squared_error(y_test,y_pred)
rmse = math.sqrt(mse)
print(rmse)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    model,
    x,
    y,
    cv=5,
    scoring="r2"
)

print(cv_scores)
print("Average CV Score:", cv_scores.mean())
```

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=1.0)

lasso.fit(x_train,y_train)

y_pred_lasso = lasso.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred_lasso))
print(mean_squared_error(y_test,y_pred_lasso))
print(r2_score(y_test,y_pred_lasso))
```

```python
lasso.coef_
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic = ElasticNet(alpha=1.0,l1_ratio=0.5)

elastic.fit(x_train,y_train)

y_pred_elastic = elastic.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred_elastic))
print(mean_squared_error(y_test,y_pred_elastic))
print(r2_score(y_test,y_pred_elastic))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100]
}

ridge_grid = GridSearchCV(
    estimator=ridge,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

ridge_grid.fit(x_train, y_train)

print("Best Alpha:", ridge_grid.best_params_)
print("Best CV Score:", ridge_grid.best_score_)

best_ridge = ridge_grid.best_estimator_

y_pred_ridge = best_ridge.predict(x_test)
```

```python
print("MAE :", mean_absolute_error(y_test, y_pred_ridge))
print("MSE :", mean_squared_error(y_test, y_pred_ridge))
print("RMSE:", mean_squared_error(y_test, y_pred_ridge)**0.5)
print("R²  :", r2_score(y_test, y_pred_ridge))
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(max_iter=10000)

param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100]
}

lasso_grid = GridSearchCV(
    estimator=lasso,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

lasso_grid.fit(x_train, y_train)

print("Best Alpha:", lasso_grid.best_params_)
print("Best CV Score:", lasso_grid.best_score_)

best_lasso = lasso_grid.best_estimator_

y_pred_lasso = best_lasso.predict(x_test)
```

```python
print("MAE :", mean_absolute_error(y_test, y_pred_lasso))
print("MSE :", mean_squared_error(y_test, y_pred_lasso))
print("RMSE:", mean_squared_error(y_test, y_pred_lasso)**0.5)
print("R²  :", r2_score(y_test, y_pred_lasso))
```

```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(max_iter=10000)

param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10],
    'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
}

elastic_grid = GridSearchCV(
    estimator=elastic,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

elastic_grid.fit(x_train, y_train)

print("Best Parameters:", elastic_grid.best_params_)
print("Best CV Score:", elastic_grid.best_score_)

best_elastic = elastic_grid.best_estimator_

y_pred_elastic = best_elastic.predict(x_test)
```

```python
print("MAE :", mean_absolute_error(y_test, y_pred_elastic))
print("MSE :", mean_squared_error(y_test, y_pred_elastic))
print("RMSE:", mean_squared_error(y_test, y_pred_elastic)**0.5)
print("R²  :", r2_score(y_test, y_pred_elastic))
```
